```python
#pakage import

import numpy as np  # ndarray
import pandas as pd  # Series, DataFrame
import matplotlib.pyplot as plt  # 시각화(그래프)
import seaborn as sns  # 시각화, 예제 데이터
```

```python
tips = sns.load_dataset('tips')
```

#Box plot

```python
# 성별 팁
sns.boxplot(data=tips, x='sex', y='tip')
plt.show()
```

```python
sns.boxplot(data=tips, x='tip', y='sex')
plt.show()
```

```python
sns.boxplot(data=tips, x='day', y='tip')
plt.show()
```

* subplot 2개를 생성
* 각각의 플롯에 흡연여부별 tips의 box plot, 시간별 box plot을 그리세요.

```python
fig, ax = plt.subplots(ncols=2, figsize=(8,4))
sns.boxplot(data=tips, x='smoker', y='tip', ax=ax[0])
sns.boxplot(data=tips, x='time', y='tip', ax=ax[1])
plt.show()
```

# 막대 그래프
## `countplot()` 함수
카테고리 타입 변수의 빈도수를 막대로 표현한 그래프


```python
# 성별 빈도수 세로 막대 그래프
sns.countplot(data=tips, x='sex')
plt.show()
```

```python
sns.countplot(data=tips, y='sex')
plt.show()
```

```python
sns.countplot(data=tips, x='day')
plt.show()
```

```python
#요일별 성별 빈도수
tips.groupby(['day','sex']).tip.count()
```

```python
sns.countplot(data=tips, x='day', hue='sex')
# hue 파라미터: 그룹(카테고리) 별로 색을 다르게 하
plt.show()
```

```python
sns.countplot(data=tips, x='sex', hue='day')
plt.show()
```

```python
# 성별, 흡연여부별 빈도수
sns.countplot(data=tips, x='sex', hue='smoker')
plt.show()
```

## `barplot()` 함수
* 그룹(카테고리) 별로 변수의 통계량(평균, 표준편차, 중앙값, ...)을 막대 그래프로 표현
* 통계량 계산 함수의 기본값은 평균(mean).


```python
#성별 팁의 평균
sns.barplot(data=tips, x='sex', y='tip')
```

```python
sns.barplot(data=tips, x='sex', y='tip', estimator='median')
```

```python
#흡연여부별 팁의 평균 시각화
sns.barplot(data=tips, x='smoker', y='tip')
```

```python
#성별 흡연 여부별 어찌구
sns.barplot(data=tips, x='smoker', y='tip', hue='sex')
```

#카테고리 그래프: `catplot()`
카테고리 별로 subplot들을 생성하고 통계값(평균, 중앙값, ...) 계산해서 각각의 subplot에 같은 종류의 그래프들을 시각화

```python
from scipy.sparse import data
# 요일별 팁 평균 막대 그래프, 흡연 여부별 팁 평균
sns.catplot(data=tips, x='day', y='tip', col='smoker', kind='bar')
```

```python
# 요일별 팁 평균 막대 그래프 시간별 서브플롯 생성
sns.catplot(data=tips, x='day', y='tip', col='time', kind='bar')
plt.show()
```

```python
# 요일별 팁 평균 막대 흡연, 시간
sns.catplot(data=tips, x='day', y='tip', col='smoker', kind='bar', row='time')
plt.show()
```

```python
# 요일별 성별 팁 막대 시간별
sns.catplot(data=tips, x='day', y='tip', hue='sex', col='time', kind='bar')
plt.show()
```

```python
# 성별 팁의 박스, 요일별 sub
sns.catplot(data=tips, x='sex', y='tip', kind='box', col='day')
plt.show()
```
