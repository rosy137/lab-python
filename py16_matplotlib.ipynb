# `matplotlib.pyplot` 모듈을 사용한 시각화

1.  `figure` 객체 생성.
1.  `figure` 객체에 subplot(axis)을 생성해서 추가.
1.  subplot에 그래프 그림.
1.  그래프의 여러가지 옵션들을 설정.
1.  그래프 보여주기.

```python
# 1. figure 객체 생성
figure = plt.figure()

# 2. 생성된 figure 객체에 subplot(axis)을 추가
# axis = figure.add_subplot(1, 1, 1)  # row, column, index
axis = figure.add_subplot(111)

# 3. subplot에 그래프 종류를 선택해서 그림
axis.hist(x=tips['total_bill'])

# 4. 그래프의 옵션 설정
axis.set_xlabel('Total Bill')
axis.set_ylabel('Count')
axis.set_title('Histogram')
axis.grid()

# 5. 그래프 보여주기
plt.show()
```

figure에 2개의 subplot을 상하로 배치하고 그래프 그리기

```python
# 1. figure 생성
fig = plt.figure(figsize=(6, 8))

# 2. subplot을 figure에 추가
ax1 = fig.add_subplot(2, 1, 1)  # add_subplot(211)
ax2 = fig.add_subplot(2, 1, 2)  # add_subplot(212)

# 3. 그래프
ax1.hist(x=tips.total_bill)
ax2.scatter(x=tips.total_bill, y=tips.tip)

# 4. 옵션
ax1.grid()
ax2.grid()

# 5. 보여주기
plt.show()
```

figure에 2개의 subplot 좌우로 배치하고 hist, scatter 그래프를 그리기

```python
fig = plt.figure(figsize=(12, 4))

ax1 = fig.add_subplot(1, 2, 1)  # add_subplot(121)
ax2 = fig.add_subplot(1, 2, 2)  # add_subplot(122)

ax1.hist(x=tips.tip)
ax2.scatter(x=tips.total_bill, y=tips.tip)

ax1.grid()
ax1.set_xlabel('Tip')
ax1.set_ylabel('Count')
ax1.set_title('Histogram')

ax2.grid()
ax2.set_xlabel('Total bill')
ax2.set_ylabel('Tip')
ax2.set_title('Scatter plot')

plt.show()
```

# `plt.subplots()` 함수

figure 생성 + subplot 추가

## figure에 1개의 subplot이 있는 경우

```python
# figure 생성, figure에 subplot을 추가.
fig, ax = plt.subplots()

# subplot에 그래프 그림.
ax.hist(x=tips.total_bill)

# 옵션

# 그래프 보여주기
plt.show()
```

## subplot 2개를 상하로 배치

```python
fig, ax = plt.subplots(nrows=2, figsize=(6, 8))
# print(ax)  # subplot이 2개 이상인 경우 리턴 값 ax는 subplot들의 배열(ndarray)

ax[0].hist(x=tips.total_bill)
ax[1].scatter(x=tips.total_bill, y=tips.tip)

plt.show()
```

## 2개의 subplot을 좌우로 배치

```python
fig, ax = plt.subplots(ncols=2, figsize=(12, 4))
# print(ax)

ax[0].hist(x=tips.total_bill)
ax[1].scatter(x=tips.total_bill, y=tips.tip)

plt.show()
```

# `pyplot` 모듈의 top-level 함수

figure 생성 + subplot 1개 생성 + 그래프 종류

## Histogram

*  연속형 변수를 구간으로 나눠서, 그 구간 안에 포함된 자료의 개수를 막대의 높이로 표현한 그래프.
*  연속형 변수의 분포를 알 수 있음.


```python
# 1. figure 생성.  2. subplot 1개 생성.  3. histogram 그림.
plt.hist(x=tips.total_bill, color='lightgray', edgecolor='black', bins=10)

# 4. 옵션 설정
plt.xlabel('Total bill')
plt.ylabel('Count')
plt.title('Histogram')
plt.grid()

# 5. 보여주기
plt.show()
```

## Bar plot(막대 그래프)

카테고리 타입 변수의 빈도수를 막대로 표현한 그래프.

```python
gender_count = tips.sex.value_counts()
gender_count
```

```python
type(gender_count)
```

```python
gender_count.index
```

```python
gender_count.values
```

```python
# 세로 막대 그래프(vertical bar plot)
# x: 인덱스, height: 막대의 높이. 빈도수.
plt.bar(x=gender_count.index, height=gender_count.values, width=0.5)

plt.show()
```

```python
# 가로 막대 그래프(horizontal bar graph)
plt.barh(y=gender_count.index, width=gender_count.values, height=0.5)

plt.show()
```

### Ex 5. 시간별 tip의 box plot을 하나의 subplot에 그리기.

```python
sns.boxplot(data=tips, x='time', y='tip')
```

```python
times = tips.time.unique()
tip_by_time = []  # Lunch, Dinner 부분집합에서 tip 컬럼을 저장하기 위한 리스트
for x in times:
    subset = tips[tips.time == x].tip
    tip_by_time.append(subset)

plt.boxplot(x=tip_by_time, labels=times)
plt.show()
```

##산점도(Scatter plot)
두 변수(컬럼) 사이의 상관 관계를 파악하기 위해서 사용


```python
#tip ~ total_bill 상관관계
plt.scatter(x=tips.total_bill, y=tips.tip,
            alpha=0.3, # 0(투명) ~ 1.0(불투명)
            s=20,      # size
            c='purple',# color
            marker='*' # shape
            )
plt.grid()
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```

```python
#tip ~ size 상관관계
plt.scatter(x=tips['size'], y=tips.tip,
            alpha=0.3, # 0(투명) ~ 1.0(불투명)
            s=20,      # size
            c='purple',# color
            marker='*' # shape
            )
plt.grid()
plt.xlabel('Size')
plt.ylabel('Tip')
plt.show()
```

##선그래프(Line plot)
주로 시계열 데이터(시간에 따라서 변하는 데이터: 주가, 환율, 날씨, ...)를 시각화

```python
np.random.seed(1)
df = pd.DataFrame(data={
    'time': np.arange(1,11), #[1,2,3, ..., 9,10]
    'price': np.random.rand(10) #[0,1.0) 범위의 난수 10개를 갖는 1차원 배열
})
df
```

```python
# price ~ time
plt.plot(df.time, df.price, 'go--')
plt.grid()
plt.xlabel('time')
plt.ylabel('price')
plt.show()
```

```python

# variable-length keyword argument(**kwargs) 사용 방법:
plt.plot(df.time, df.price,
         c='crimson',  # color
         marker='*',   # marker
         linestyle='dotted'  # line style
         )

plt.show()
```

#Ex 3.
* tip ~ total_bill의 상관관계를 scatter plot으로 그리세요.
* 요일별로 점의 색깔, 모양을 다르게 표현하세요.
* 범례를 표시.

```python
days=tips.day.unique()
markers = ['o','v','s','*']
colors = ['pink','orange','green','blue']
for d,m,c in zip(days,markers,colors):
    plt.scatter(x = tips[tips.day==d].total_bill, y = tips[tips.day==d].tip,
                alpha=0.3,
                s=20,
                c=c,
                marker=m,
                label=d
                )
plt.grid()
plt.legend()
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```
